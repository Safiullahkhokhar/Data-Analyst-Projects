<font size='8'>***Scrape Channel statistics from youtube.***</font>

```python
# channel details
def get_channel_stats(youtube, channel_ids):
    
            # youtube.channels() it is a resource name 
    request = youtube.channels().list(
            part= 'snippet,contentDetails,statistics',
            id = ','.join (channel_ids) ) # list convert into a string where all the values are comma seprated
    
    response = request.execute()
    all_data = []
    for i in range(len(response['items'])):
        # access the data which is needed
        data = dict( Channel_name = response['items'][i] ['snippet']['title'],
                    Subscribers = response['items'][i] ['statistics']['subscriberCount'],
                    Views = response['items'][i] ['statistics']['viewCount'],
                    Total_video = response['items'][i] ['statistics']['videoCount'],
                    Playlist_ids = response['items'][i] ['contentDetails']['relatedPlaylists']['uploads'])
        all_data.append(data)
        
    return all_data
```

<font size = '5'>***Video ids***</font>

```python
# function to get video ids
def get_video_ids(youtube, Playlist_ids_var):
    request = youtube.playlistItems().list(
                part = 'contentDetails',
                playlistId = Playlist_ids_var, 
                maxResults = 50)# it extract 50 max results
    response = request.execute()

    videos_ids=[]
    # itreate on video ids
    for i in range(len(response['items'])):
        videos_ids.append(response['items'][i] ['contentDetails']['videoId'])

    # next page tokens
    next_page_token = response.get('nextPageToken')
    more_pages = True # contains more page then true

    while more_pages:
        # if next_page_token can't cantain tokens than more pages will be false
        if next_page_token is None:
            more_pages = False
        else:
            request = youtube.playlistItems().list(
                part = 'contentDetails',
                playlistId = Playlist_ids_var, 
                maxResults = 50, # it extract 50 max results
                pageToken = next_page_token)
            response = request.execute()

            for i in range(len(response['items'])):
                videos_ids.append(response['items'][i] ['contentDetails']['videoId'])
                 
            next_page_token = response.get('nextPageToken')



    return videos_ids
    #return len(videos_ids) # this is the videos_ids list which contain 50 of that channel
                           # if the user can return the list then cal "return videos_ids" 
```

```python
# cal the video ids
video_ids = get_video_ids(youtube, Playlist_ids_var)
video_ids # total no. of videos
```

<font size = '5'>***Video Details***</font>

```python
# function to get video details
def get_video_details(youtube, video_ids):
    all_video_stats=[]
    # itrate on total videos
    for i in range(0, len(video_ids), 50): # start with 0 untill the total no. of videos(644) but it takes 50 at a time 
        request = youtube.videos().list(
            part = "snippet, statistics",
            id = ','.join(video_ids [ i : i+50 ]))# i=50: 50+50 untill total no. of videos(644)
        response = request.execute()# response object will have deatils for 50 videos
        # in this loop through 50 times to get details of each video
        for video in response['items']:
            # all the videos details is stored in a dictionary
            video_stats = dict(Title = video['snippet']['title'],
                               Published_date = video['snippet']['publishedAt'],
                               Views = video['statistics']['viewCount'],
                               Likes = video['statistics']['likeCount'],
                               # Dislikes =  video['statistics']['dislikeCount'],
                               Comments = video['statistics']['commentCount'],)
            all_video_stats.append(video_stats)
    
    # return len(all_video_stats) #644
    return all_video_stats
```

```python
# call the function
video_details = get_video_details(youtube, video_ids)
video_details
```
