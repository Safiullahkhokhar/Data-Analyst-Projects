```python
# import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# import the data file 
df = pd.read_csv('Sales Data.csv', encoding='unicode_escape')
# to avoid encoding error, use 'unicode_escape'
```

```python
# drop missing data
df.drop(['Status', 'unnamed1'], axis=1, inplace=True) 
```

```python
# check null values
df.isnull().sum()
```

```python
df.shape
```

```python
# droupna cleaning up DataFrames with missing values. 
# Used to remove rows/columns that are not complete
df.dropna(inplace=True)
df.shape
```

```python
# change data type
df['Amount'] = df['Amount'].astype('int')
```

<font size='5'>***Gender***</font>

```python
sns.countplot( x='Gender', data= df, width = 0.5, palette='hls')
```

```python
ax = sns.countplot( x='Gender', data= df, width= 0.5, palette='hls')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Comparison amount on gender bases male and female 
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data= sales_gen, width= 0.5, palette='hls')
```

From the above graphs we can see that most of the buyers are females and even the purchesing power of females are greater the man

<font size='5'>***Age***</font>

```python
df.columns
```

```python
ax = sns.countplot( data= df, x='Age Group', palette='hls')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Comparison total amount and age group 
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data= sales_age, palette='hls')
```

From above graph we can see that most of the buyers are of age group between 26-35yrs female

<font size='5'>***State***<font>

```python
df.columns
```

```python
# Total no. of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by= 'Orders', ascending=False).head(10)

# size of the plot
sns.set( rc={'figure.figsize' : (16,4)})
# total orders
sns.barplot( data= sales_state, x= 'State', y= 'Orders', palette='hls')
```

```python
# Total amount from top 10 states 
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False).head(10)

# size of the plot
sns.set( rc={'figure.figsize' : (16,4)})
# total amount 
sns.barplot( data= sales_state, x= 'State', y= 'Amount', palette='hls')
```

From aboves graphs we can see that unexpectedly most of the orders are form Uttar Pradesh, Maharashtra and Kanateka respectively
but total sales amount is from UP, Kanateka and Maharashtra

```python
ax = sns.countplot( data= df, x='Marital_Status', width= 0.5, palette='hls')

sns.set( rc={'figure.figsize' : (7,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

df.columns

```python
sns.set( rc={'figure.figsize' : (21,5)})

# palette = {"A": "red", "B": "green", "C": "blue"}
ax = sns.countplot( data= df, x='Occupation', palette='hls')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot( data= df, x='Product_Category', palette='hls')

sns.set( rc={'figure.figsize' : (27,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```
